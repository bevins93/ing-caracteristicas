```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
# Start coding here
# Use as many cells as you need
```

```python
# Frecuencia de los delitos por hora
crimes['Hour'] = crimes['TIME OCC'].str[:2].astype(int)
crimeByHour = crimes['Hour'].value_counts()

# Hora con mayor frecuencia de delitos
peak_crime_hour = crimeByHour.idxmax()
peak_crime_hour
```

```python
# Delitos cometidos entre las 22.00 y las 4.00 horas
night_crimes = crimes[(crimes['Hour'] >= 22) | (crimes['Hour'] <= 4)]

# Frecuencia de los delitos nocturnos por zona
night_crime_by_area = night_crimes['AREA NAME'].value_counts()

# Zona con mayor frecuencia de delitos nocturnos
peak_night_crime_location = night_crime_by_area.idxmax()
peak_night_crime_location
```

```python
# Calcular los grupos de edad
age_groups = ['<18', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']

# Nueva columna 'Age Group' basada en la edad de la vÃ­ctima
crimes['Age Group'] = pd.cut(crimes['Vict Age'], bins=[0, 18, 25, 34, 44, 54, 64, np.inf], labels=age_groups)

# NÃºmero de delitos por grupo de edad
victim_ages = crimes['Age Group'].value_counts()
victim_ages
```
